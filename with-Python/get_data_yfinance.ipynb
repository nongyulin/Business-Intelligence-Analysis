# Examples of Obtaining Data
yfinance can be used to fetch YaHoo!finance data like bond data(1), stock indices(2), company financial and non-financial data(3), commodities(4), currencies(5), international markets(6), and crypto-currency(7).

```python
target = ["^TNX", "^IXIC", "BILI","CL=F","EURUSD=X","^N225", "BTC-USD"]
```

## 1 Getting Bond Data : 10-Yr Bond

```python
yb_10 = yf.Ticker(target[0])
yb_10.info
```

```python
yb_10.history(period='5D')
```
