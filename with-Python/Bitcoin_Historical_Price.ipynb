# Data Visualisation 
There are 4 kinds of chart been used frequently, which are Line Chart, Bar Chart, Candlestrik Chart, Renko Chart, and Point and Figure Chart.
* **Line chart** is a simple graphic display of price trends over time. It has one data point per time interval.
* **Bar chart** has four bits of data in each entry, which are the high and low price encountered during the imte interval, the opening and closing prices. From bar chart, analysts can immediately gets a sense of the nature of the time interval's trading. 
* **Candlestick chart** also provides four prices per data point entry : the high and low prices, and the opening and closing prices during the period. The body of the candle is shaded or clear due to price difference of opening and closing. Compared to Bar Chart, Candlestick Chart is more visible.
* **Renko chart** looks like a series of bricks. A new brick is created when the price moves a specified price amount.
* **Point and figure chart** was widely used in US in the early 1900s. Neither time nor volume are represented in the chart. To draw the point and figure chart, we have to decide the box size and the reversal size.
