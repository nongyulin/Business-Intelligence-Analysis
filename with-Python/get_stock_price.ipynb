# Get from Yahoo Finance 

```python
import yfinance as yf
import finplot as fplt

df_yh = yf.download('TSLA',start='2019-01-01', end = '2020-04-29')
df_yh.head()
```

```python
df_yh.loc['2019-01-02']
```

```python
df_yh.info()
```

```python
df_yh.describe().T
```

```python
fplt.candlestick_ochl(df_yh[['Open','Close','High','Low']])
fplt.plot(df_yh.Close.rolling(5).mean())
fplt.plot(df_yh.Close.rolling(10).mean())
fplt.show()
```
