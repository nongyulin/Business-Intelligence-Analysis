# Import Needed Package

```python
import requests
from bs4 import BeautifulSoup
import json
import pandas as pd
```

```python
Financial_url=pd.DataFrame()
Financial_url['Income Statement']=['https://finance.yahoo.com/quote/TSLA/financials?p=TSLA']
Financial_url['Balance Sheet']=['https://finance.yahoo.com/quote/TSLA/balance-sheet?p=TSLA']
Financial_url['Cash Flow']=['https://finance.yahoo.com/quote/TSLA/cash-flow?p=TSLA']
Financial_url
```

# Get Financial Report
## Income Statement

```python
response_IS = requests.get(Financial_url.loc[0][0])
soup_IS = BeautifulSoup(response_IS.text, 'html.parser')

div_IS = soup_IS.find_all('div', class_ = "D(tbr) C($primaryColor)")[0].find_all('span')


columns_IS = [span.text for span in div_IS[1:]]
div_IS = soup_IS.find_all('div', class_ = 'D(tbrg)')[0]
values_IS = dict()
for element in div_IS:
  try:
    values_IS[element.find_all('div', class_ = "D(ib) Va(m) Ell Mt(-3px) W(215px)--mv2 W(200px) undefined")[0].get('title')] = [i.text for i in element.find_all('div', {'data-test' : 'fin-col'})]
  except:
    pass
print(pd.DataFrame(values_IS)) 
```

```python
IncomeStatement=pd.DataFrame(values_IS)
IncomeStatement
```

## Balance Sheet 

```python
response_BS = requests.get(Financial_url.loc[0][1])
soup_BS = BeautifulSoup(response_BS.text, 'html.parser')

div_BS = soup_BS.find_all('div', class_ = "D(tbr) C($primaryColor)")[0].find_all('span')

columns_BS = [span.text for span in div_BS[1:]]
div_BS = soup_BS.find_all('div', class_ = 'D(tbrg)')[0]
values_BS = dict()
for element in div_BS:
  try:
    values_BS[element.find_all('div', class_ = "D(ib) Va(m) Ell Mt(-3px) W(215px)--mv2 W(200px) undefined")[0].get('title')] = [i.text for i in element.find_all('div', {'data-test' : 'fin-col'})]
  except:
    pass
print(pd.DataFrame(values_BS)) 
```

```python
BalanceSheet=pd.DataFrame(values_BS)
BalanceSheet
```

```python
response_CF = requests.get(Financial_url.loc[0][2])
soup_CF = BeautifulSoup(response_CF.text, 'html.parser')

div_CF = soup_CF.find_all('div', class_ = "D(tbr) C($primaryColor)")[0].find_all('span')


columns_CF = [span.text for span in div_CF[1:]]
div_CF = soup_CF.find_all('div', class_ = 'D(tbrg)')[0]
values_CF = dict()
for element in div_CF:
  try:
    values_CF[element.find_all('div', class_ = "D(ib) Va(m) Ell Mt(-3px) W(215px)--mv2 W(200px) undefined")[0].get('title')] = [i.text for i in element.find_all('div', {'data-test' : 'fin-col'})]
  except:
    pass
print(pd.DataFrame(values_CF)) 
```

```python
CashFlow=pd.DataFrame(values_CF)
CashFlow
```
